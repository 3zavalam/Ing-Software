```python
import pandas as pd
```

```python
def normalize_column_names(df):
    # Primero: corregir sufijos automáticos .1, .2 → _1, _2
    new_cols = []
    for col in df.columns:
        if '.' in col:
            base, suffix = col.split('.', 1)
            if suffix.isdigit():
                new_cols.append(f"{base}_{suffix}")
            else:
                new_cols.append(col)
        else:
            new_cols.append(col)
    df.columns = new_cols

    # Segundo: renombrar duplicados si aún existen
    seen = {}
    final_cols = []
    for col in df.columns:
        if col not in seen:
            seen[col] = 0
            final_cols.append(col)
        else:
            seen[col] += 1
            final_cols.append(f"{col}_{seen[col]}")
    df.columns = final_cols

    return df
```

```python
players_mx = normalize_column_names(pd.read_csv("data/radar/Liga_MX/players.csv"))
keepers_mx = normalize_column_names(pd.read_csv("data/radar/Liga_MX/keepers.csv"))
playersMLS = normalize_column_names(pd.read_csv("data/radar/MLS/players.csv"))
keepersMLS = normalize_column_names(pd.read_csv("data/radar/MLS/keepers.csv"))
```

```python
# Guardar DataFrames corregidos
players_mx.to_csv("data/radar/Liga_MX/players.csv", index=False)
keepers_mx.to_csv("data/radar/Liga_MX/keepers.csv", index=False)

playersMLS.to_csv("data/radar/MLS/players.csv", index=False)
keepersMLS.to_csv("data/radar/MLS/keepers.csv", index=False)

print("✅ Archivos CSV guardados con columnas normalizadas.")
```
